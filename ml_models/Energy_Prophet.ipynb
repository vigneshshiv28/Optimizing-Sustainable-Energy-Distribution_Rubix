```python
import pandas as pd
import numpy as np
from prophet import Prophet
```

```python
# df = pd.read_csv('/content/DL_Complete_Yearly_Demand_Profile_train.csv')
df = pd.read_csv('/content/DL_Yearly_Demand_Profile_train.csv')
df = df.drop('Unnamed: 0',axis =1)
df.head()
```

```python
# df = df.rename(columns={'DateTime': 'ds', 'Hourly Demand Met (in MW)': 'y'})

df['y'] = pd.to_numeric(df['y'])
df['ds'] = pd.to_datetime(df['ds'])
df.head()
```

```python
df.ds.dtype
```

```python
m = Prophet()
# m.add_regressor('Temperature')
# m.add_regressor('Humidity')
m.fit(df)
```

```python
# temp = np.random.uniform(9, 11, 857)
# humidity = np.random.uniform(50, 100, 857)
```

```python
future = m.make_future_dataframe(periods=7*24,freq='H')

# future['Temperature'] = temp
# future['Humidity'] = humidity
```

```python
forecast = m.predict(future)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()

```
